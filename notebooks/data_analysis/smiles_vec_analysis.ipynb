PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=8)

principalComponents = pca.fit_transform(smiles_vec)
data = pd.DataFrame(data = principalComponents)
```

```python
data.head()
```

```python
data = (data.subtract(data.mean(axis=0), axis=1).div(data.std(ddof=0, axis=0), axis=1))
data = data.div(data.abs().max(axis=0), axis=1)
```

```python
data.head()
```

```python
indexed_data = pd.concat([smiles_vec.reset_index()["0"], data], axis =1, ignore_index=True).rename(columns={0: 'smiles'}).set_index('smiles')
```

```python
transposed = indexed_data.transpose()
transposed[' '] = 0.0000000001
indexed_data = transposed.transpose()
```

```python
smiles_dict = pd.DataFrame.from_dict(SMILES_CHARACTER_TO_ID, orient='index')
```

```python
merged_data = pd.merge(right=indexed_data, left=smiles_dict, how="inner",left_index=True, right_index=True)
```

```python
final = merged_data.values
```
